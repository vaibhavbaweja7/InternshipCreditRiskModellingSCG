### Then,you need to import the matplotlib library into your Python code.

```python
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
```

# Plotting a scatter plot using Matplotlib


```python
# Load the data into a Pandas DataFrame.
data = pd.read_csv(r"C:\Users\vaibh\Desktop\car_data.csv")
# Create a scatter plot of Selling_Price vs. Kms_Driven
plt.scatter(data['Selling_Price'], data['Kms_Driven'])

# Add a title and labels to the axes
plt.title('Selling Price vs. Kms Driven')
plt.xlabel('Selling Price')
plt.ylabel('Kms Driven')

# Show the plot
plt.show()
```

# Plotting the bar chart using Matplotlib

```python

# Create a bar chart of Fuel_Type vs. Selling_Price
plt.bar(data['Fuel_Type'], data['Selling_Price'])

# Add a title and labels to the axes
plt.title('Selling Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Selling Price')

# Show the plot
plt.show()
```

# Plotting the pie chart using Matplotlib

```python
# Get the data for the pie chart
fuel_types = data['Fuel_Type'].unique()
counts = data['Fuel_Type'].value_counts()

# Plot the pie chart
plt.pie(counts, labels=fuel_types)

# Add a title to the pie chart
plt.title('Cars by Fuel Type')

# Show the pie chart
plt.show()
```

# Plotting the line plot for one car

```python
# Filter the data to only include Ritz cars
ritz_data = data[data['Car_Name'] == 'ritz']
ritz_data= ritz_data.sort_values(by= 'Year')

# Create a line plot
plt.plot(ritz_data['Year'], ritz_data['Selling_Price'], label='Ritz Car')

# Set the title, x-axis label, and y-axis label
plt.title('Ritz Car Selling Price')
plt.xlabel('Year')
plt.ylabel('Selling Price')

plt.show()
```

#  Plotting a Seaborn Scatter plot

```python

# Create a Seaborn figure and axes
fig, ax = plt.subplots()

# Plot the data using the Seaborn `scatterplot()` function
sns.scatterplot(x='Year', y='Selling_Price', data=data, ax=ax)

# Add labels to the axes and title to the plot
ax.set_xlabel('Year')
ax.set_ylabel('Selling_Price')
ax.set_title('Car Selling Price by Year')

# Show the plot
plt.show()
```

# Heatmap of Correlation using seaborn

```python
# Plot the heatmap
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
```
